```
import numpy as np
import cv2
from skimage import io
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow

def url_to_image(url):
  print("downloading %s" % (url))
  return cv2.cvtColor(io.imread(url), cv2.COLOR_BGR2RGB)

debug = False

def grab_foreground(url, sp, sr, rect_x, rect_y, rect_w, rect_h):
  img = url_to_image(url)

  cv2_imshow(img)

  shifted = cv2.pyrMeanShiftFiltering(img, sp, sr)
  if debug == True:
    print("After filtering")
    cv2_imshow(shifted)

  (w,h) = img.shape[:2]
  mask = np.zeros((w, h), np.uint8)
  bgdModel = np.zeros((1, 65),np.float64)
  fgdModel = np.zeros((1, 65),np.float64)

  cv2.grabCut(shifted, mask, (rect_x, rect_y, rect_w, rect_h), bgdModel, fgdModel, 2, cv2.GC_INIT_WITH_RECT)
  mask2 = np.where((mask==2)|(mask==0), 0, 1).astype('uint8')
  img = img*mask2[:,:,np.newaxis]
  if debug == True:
    print("Mask")
    cv2_imshow(mask2*255)

  cv2_imshow(img)





  
```

```
url = "https://raw.githubusercontent.com/mdavydov/ComputerVisionCourse/master/tasks/object/obj1.jpg"
grab_foreground(url, sp=20, sr=30, rect_x=30, rect_y=30, rect_w=430, rect_h=220)
```
